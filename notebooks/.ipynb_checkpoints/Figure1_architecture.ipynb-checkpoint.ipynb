```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

plt.figure(figsize=(6, 4))
ax = plt.gca(); ax.axis('off')

boxes = [("L3\nStoryline Buffer", 0.1, 0.6),
         ("L2\nScene Graph",      0.1, 0.35),
         ("L1\nSensory Maps",     0.1, 0.1)]

for text, x, y in boxes:
    ax.add_patch(patches.FancyBboxPatch((x, y), 0.8, 0.18,
                                        boxstyle="round,pad=0.02"))
    ax.text(x + 0.4, y + 0.09, text, ha='center', va='center')

plt.savefig("figure1_architecture.png", dpi=300)
plt.show()
```
