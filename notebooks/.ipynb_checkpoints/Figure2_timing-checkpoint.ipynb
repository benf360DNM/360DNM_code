```python
# Figure 2 – Five-step 100 ms generative loop timeline
import matplotlib.pyplot as plt

# Step labels and their order
steps = ["Prediction", "Observation", "Error", "Update", "Broadcast"]
x = list(range(len(steps)))          # [0, 1, 2, 3, 4]

plt.figure(figsize=(7, 2))
plt.plot(x, [0]*len(x), marker='o')  # a simple horizontal line with dots
plt.yticks([])                       # hide the y-axis

# Add angled text labels under each dot
for idx, label in enumerate(steps):
    plt.text(idx, -0.1, label, rotation=45, ha='right', va='top')

plt.xlabel("Cycle step (total cycle ≈ 100 ms)")
plt.title("Figure 2 – Five-step Generative Loop Timeline")
plt.tight_layout()

# Save the figure in the project root (same folder as Figure 1 image)
plt.savefig("figure2_timing.png", dpi=300)
plt.show()
```
