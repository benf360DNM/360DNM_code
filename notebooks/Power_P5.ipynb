**Variable:** parietal → visual β-band gain correlates r > 0.5 with subjective continuity ratings  
**Paradigm:** MEG + immersive VR storyline


```python
# Power P5 – simulate β-band gain vs continuity
import sys, numpy as np, matplotlib.pyplot as plt, scipy.stats as stats
sys.path.append("..")

from src.power_analysis import simulate_p5     # returns (beta_gain, continuity)

gain, cont = simulate_p5()
r, _       = stats.pearsonr(gain, cont)

print(f"Correlation r = {r:.2f}   (criterion r > 0.50)")

plt.figure(figsize=(6,4))
if gain.size > 1:
    plt.scatter(gain, cont, color="lightblue", edgecolor="k")
    fit = np.poly1d(np.polyfit(gain, cont, 1))(gain)
    plt.plot(gain, fit, color="r")
else:
    plt.scatter([0], [cont], color="lightblue", edgecolor="k")
plt.xlabel("β-band gain")
plt.ylabel("Continuity rating")
plt.title("β-band feedback vs continuity")
plt.tight_layout()
plt.savefig("figure_P5_beta_gain.png", dpi=300)
plt.show()
```
