# Power P6 – Theta phase coupling and foresight  
**Variable:** 4–8 Hz hippocampus–angular-gyrus phase coherence predicts ≥ 30 % of trial-level vividness variance  
**Paradigm:** stereo-EEG in epilepsy patients + episodic future-thinking task


```python
# Power P6 – simulate θ-phase coherence vs vividness
import sys, numpy as np, matplotlib.pyplot as plt, scipy.stats as stats
sys.path.append("..")

from src.power_analysis import simulate_p6     # returns (theta_coh, vividness)

coh, vivid = simulate_p6()
r, _       = stats.pearsonr(coh, vivid)
r2         = r**2 * 100

print(f"Variance explained = {r2:.1f}%   (criterion ≥ 30 %)")

plt.figure(figsize=(6,4))
if coh.size > 1:
    plt.scatter(coh, vivid, color="gold", edgecolor="k")
    fit = np.poly1d(np.polyfit(coh, vivid, 1))(coh)
    plt.plot(coh, fit, color="r")
else:
    plt.scatter([0], [vivid], color="gold", edgecolor="k")
plt.xlabel("θ-phase coherence")
plt.ylabel("Vividness")
plt.title("θ-phase coupling vs vividness")
plt.tight_layout()
plt.savefig("figure_P6_theta_vividness.png", dpi=300)
plt.show()
```
