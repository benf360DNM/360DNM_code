```python
# Power P3 – simulate dream-coherence change with replay acceleration
import sys, numpy as np, matplotlib.pyplot as plt
sys.path.append("..")

from src.power_analysis import simulate_p3   # returns (accel_pct, coherence)

accel, coh = simulate_p3()
slope, _   = np.polyfit(accel, coh, 1)

print(f"Slope β = {slope:.3f}   (criterion β > 0.4)")

plt.figure(figsize=(6,4))
if accel.size > 1:
    plt.scatter(accel, coh, color="skyblue", edgecolor="k")
    fit = np.poly1d(np.polyfit(accel, coh, 1))(accel)
    plt.plot(accel, fit, color="r")
else:
    plt.scatter([0], [coh], color="skyblue", edgecolor="k")
plt.xlabel("Replay acceleration (%)")
plt.ylabel("Dream coherence score")
plt.title("Dream-coherence vs replay acceleration")
plt.tight_layout()
plt.savefig("figure_P3_dream_coherence.png", dpi=300)
plt.show()
```
