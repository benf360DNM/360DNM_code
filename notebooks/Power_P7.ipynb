# Power P7 – Pharmacological precision damping  
**Variable:** ≥ 30 % reduction in P3b amplitude under 0.3 mg kg⁻¹ ketamine versus placebo  
**Paradigm:** pharmaco-EEG + narrative self-report


```python
# Power P7 – simulate P3b reduction under ketamine
import sys, numpy as np, matplotlib.pyplot as plt
sys.path.append("..")

from src.power_analysis import simulate_p7   # returns (placebo, ketamine)

placebo, keta = simulate_p7()
mean_placebo  = placebo.mean()
mean_keta     = keta.mean()
reduction_pct = (mean_placebo - mean_keta) / mean_placebo * 100

print(f"P3b reduction = {reduction_pct:.1f}%   (criterion ≥ 30 %)")

plt.figure(figsize=(4,4))
x = [0, 1]
plt.bar(x, [mean_placebo, mean_keta], color=["slateblue","salmon"])
plt.xticks(x, ["Placebo", "Ketamine"])
plt.ylabel("Mean P3b amplitude (µV)")
plt.title("P3b amplitude by condition")
plt.tight_layout()
plt.savefig("figure_P7_p3b_reduction.png", dpi=300)
plt.show()
```
