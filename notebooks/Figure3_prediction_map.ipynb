```python
# Figure 3 – Prediction–Paradigm map (P1–P7)
import matplotlib.pyplot as plt
import numpy as np

preds = ["P1", "P2", "P3", "P4", "P5", "P6", "P7"]
paradigms = ["EEG/MEG", "TMS", "Sleep-EEG",
             "EEG", "MEG", "Stereo-EEG", "Pharmaco-EEG"]

y = np.arange(len(preds))

plt.figure(figsize=(6, 4))
plt.scatter([1]*len(preds), y)           # x=1 for all; y positions staggered
plt.yticks(y, [f"{p}: {par}" for p, par in zip(preds, paradigms)])
plt.xticks([])                           # hide x-axis ticks
plt.title("Prediction–Paradigm Map")
plt.tight_layout()
plt.savefig("figure3_prediction_map.png", dpi=300)
plt.show()
```
