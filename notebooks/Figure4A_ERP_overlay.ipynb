```python
# Figure 4A – Simulated ERP overlay (congruent vs. scrambled)
import numpy as np
import matplotlib.pyplot as plt

# Time axis in milliseconds
t = np.linspace(0, 400, 401)

def erp_peak(t, mu, amp):
    """Gaussian-shaped ERP component"""
    return amp * np.exp(-0.5 * ((t - mu) / 20) ** 2)

# Build two synthetic waveforms
erp_congruent = (erp_peak(t, 70, 3) -               # early P1-like
                 erp_peak(t, 140, 5) +              # N1-like
                 erp_peak(t, 210, 4))               # P2/P3-like

erp_scrambled = (erp_peak(t, 80, 3) -               # shifted peaks
                 erp_peak(t, 150, 5) +
                 erp_peak(t, 230, 4))

plt.figure(figsize=(7, 3))
plt.plot(t, erp_congruent, label="Congruent")
plt.plot(t, erp_scrambled,  linestyle="--", label="Scrambled")
plt.xlabel("Time (ms)")
plt.ylabel("Amplitude (µV)")
plt.title("Figure 4A – Predicted ERP Overlay")
plt.legend()
plt.tight_layout()
plt.savefig("figure4A_erp_overlay.png", dpi=300)
plt.show()
```
